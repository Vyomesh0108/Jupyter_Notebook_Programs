```python
# By importing this, we can represent XML file as a tree, ElementTree breaks into a tree structure which is easy to work with.
import xml.etree.ElementTree as ET
```

```python
# We have passed or extracted the xml file & stored in file. Every part of a tree (root included) has a tag that describe element.
file = ET.parse('college.xml')
```

```python
root = file.getroot()
```

```python
root
```

```python
root.tag
# To print tag name
```

```python
file
# For showing the reference of XML in file format
```

```python
for node in root:
    tid = node.find('id').text
    tna = node.find('name').text
    tpe = node.find('per').text
    
    print(tid, tna, tpe)
```

```python
# For converting XML into dataframe with pandas
import pandas as pd

rows = []

for node in root:
    tid = node.find('id').text
    tna = node.find('name').text
    tpe = node.find('per').text
    
    rows.append({"id": tid, "name": tna, "per": tpe})
```

```python
df = pd.DataFrame(rows)
```

```python
df
```

```python
def change_name(na):
    return na.upper()

# For changing name to uppercase
```

```python
df['name'] = df['name'].apply(change_name)
df
```

```python
def update_result(p):
    if int(p)>60: 
        return "excellent"
    else:
        return "good"
    
# This will update result, of 'per' is greater than 60 it will show 'excellent' & less than 60, then 'good'
```

```python
df['result'] = df['per'].apply(update_result)
```

```python
import pandas as pd 

people = {
    'first': ['het', 'neel', 'bala', 'thomas', 'kiran', 'vyomesh'],
    'last': ['patel', 'alvida', 'shah', 'chipku', 'nelson', 'mistry'],
    'college': ['itm', 'bit', 'svit', 'itm', 'ntc', 'ntc'],
    'per': [70, 80, 50, 90, 60, 75],
    'result': ['a', 'a', 'a', 'a', 'a', 'a']
}

df = pd.DataFrame(people)
df
```

```python
# Replace
df['first'].replace({'het':'twfs', 'thomas':'darsh'})
```
