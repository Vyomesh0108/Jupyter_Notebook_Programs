```python
import networkx as nx
from matplotlib import pyplot as plt

G = nx.Graph();

print(G.nodes())
print(G.edges())

print(type(G.nodes()))
print(type(G.edges()))
```

```python
G = nx.Graph()

G.add_node("A")
G.add_node("B")

G.add_nodes_from(["C", "D", "E", "F"])

print("NODES = ",G.nodes())
print("EDGES = ",G.edges())

nx.draw(G, with_labels=True, node_color="red")
plt.savefig("Nodes")
```

```python
# Graph comes in straight line
G = nx.Graph()

G.add_node("A")
G.add_node("B")

G.add_nodes_from(["C", "D", "E", "F"])
G.add_edges_from([("A", "C"),("C", "E"), ("E", "B"), ("B", "F"), ("F", "D")])

print("NODES = ",G.nodes())
print("EDGES = ",G.edges())

nx.draw(G, with_labels=True, node_color="red")
plt.savefig("Nodes & Edges With Straight Line")
```

```python
# Graph comes in different shape
G = nx.DiGraph()

G.add_node("A")
G.add_node("B")

G.add_nodes_from(["C", "D", "E", "F"])
G.add_edges_from([("A", "C"),("C", "E"), ("E", "B"), ("B", "F"), ("F", "D")])

print("NODES = ",G.nodes())
print("EDGES = ",G.edges())

nx.draw(G, with_labels=True, node_color="red")
plt.savefig("Nodes & Edges With Different Shape")
```

```python
# Graph comes in different shape
G = nx.DiGraph()

G.add_node("A")
G.add_node("B")

G.add_nodes_from(["C", "D", "E", "F"])
G.add_edges_from([("A", "C"),("C", "E"), ("E", "B"), ("B", "F"), ("F", "D")])

print("NODES = ",G.nodes())
print("EDGES = ",G.edges())

colors = ['g', 'm', 'r', 'm', 'r', 'g']
size = [1000, 200, 400, 300, 600, 1200]
ed_col = ['blue', 'yellow', 'green', 'magenta', 'gray']

# shape = 'o', 'd', 'D', 's' 
nx.draw_networkx(G, node_color=colors, node_size=size, edge_color=ed_col, node_shape='s')
plt.title("Graph of College Network With Square Shape")
plt.savefig("Graph of College Network With Square Shape")
plt.show()
```

```python
G = nx.DiGraph()

G.add_node("A")
G.add_node("B")

G.add_nodes_from(["C", "D", "E", "F"])
G.add_edges_from([("A", "C"),("C", "E"), ("E", "B"), ("B", "F"), ("F", "D")])

print("NODES = ",G.nodes())
print("EDGES = ",G.edges())

G.remove_node("E")

nx.draw(G, with_labels=True, node_color="red")
plt.savefig("Nodes & Edges With One Node Remove")
```

```python
# Graph with shortest path
G = nx.DiGraph()
G.add_edge('a', 'b', weight=0.1)
G.add_edge('b', 'c', weight=1.5)
G.add_edge('a', 'c', weight=1.0)
G.add_edge('c', 'd', weight=2.2)

nx.draw(G, with_labels=True, node_color="red")
print(nx.shortest_path(G, 'b', 'd'))
plt.savefig("Graph With Shortest Path")
```
